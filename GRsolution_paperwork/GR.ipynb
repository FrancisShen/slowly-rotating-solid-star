 Given

  Metric:
  $$ds^2 = -H dt^2 + Q dr^2 + r^2 K d\theta^2 + r^2 K \sin^2\theta (d\phi - \Omega dt)^2$$

  Orthonormal coframe:
  $$\theta^0 = \sqrt{H},dt,\quad \theta^1 = \sqrt{Q},dr,\quad \theta^2 = r\sqrt{K},d\theta,\quad \theta^3 = r\sqrt{K}\sin\theta,(d\phi - \Omega
  dt)$$

  Inverse relations (coordinate → tetrad):
  $$dt = \frac{\theta^0}{\sqrt{H}},\quad dr = \frac{\theta^1}{\sqrt{Q}},\quad d\theta = \frac{\theta^2}{r\sqrt{K}},\quad d\phi = \Omega,dt +
  \frac{\theta^3}{r\sqrt{K}\sin\theta}$$

  ---
  1. Computation of $d\theta^0 = d(\sqrt{H},dt)$

  Leibniz rule:
  $$d\theta^0 = d(\sqrt{H}) \wedge dt + \sqrt{H} \wedge \underbrace{d(dt)}_{=0}$$

  Scalar differential:
  $$d(\sqrt{H}) = \frac{\partial\sqrt{H}}{\partial r}dr + \frac{\partial\sqrt{H}}{\partial\theta}d\theta = \frac{H_r}{2\sqrt{H}}dr +
  \frac{H_\theta}{2\sqrt{H}}d\theta$$

  Thus in the coordinate basis:
  $$d\theta^0 = \frac{H_r}{2\sqrt{H}},dr \wedge dt + \frac{H_\theta}{2\sqrt{H}},d\theta \wedge dt$$

  Convert to orthonormal basis:
  $$dr \wedge dt = \frac{\theta^1}{\sqrt{Q}} \wedge \frac{\theta^0}{\sqrt{H}} = \frac{1}{\sqrt{HQ}},\theta^1 \wedge \theta^0 =
  -\frac{1}{\sqrt{HQ}},\theta^0 \wedge \theta^1$$

  $$d\theta \wedge dt = \frac{\theta^2}{r\sqrt{K}} \wedge \frac{\theta^0}{\sqrt{H}} = \frac{1}{r\sqrt{HK}},\theta^2 \wedge \theta^0 =
  -\frac{1}{r\sqrt{HK}},\theta^0 \wedge \theta^2$$

  Plug in:
  $$d\theta^0 = \frac{H_r}{2\sqrt{H}}\left(-\frac{1}{\sqrt{HQ}}\right)\theta^0 \wedge \theta^1 +
  \frac{H_\theta}{2\sqrt{H}}\left(-\frac{1}{r\sqrt{HK}}\right)\theta^0 \wedge \theta^2$$

  $$\boxed{d\theta^0 = -\frac{H_r}{2H\sqrt{Q}};\theta^0 \wedge \theta^1 ;-; \frac{H_\theta}{2rH\sqrt{K}};\theta^0 \wedge \theta^2}$$

  ---
  2. Computation of $d\theta^1 = d(\sqrt{Q},dr)$

  Leibniz rule:
  $$d\theta^1 = d(\sqrt{Q}) \wedge dr + \sqrt{Q} \wedge \underbrace{d(dr)}_{=0}$$

  Scalar differential:
  $$d(\sqrt{Q}) = \frac{Q_r}{2\sqrt{Q}}dr + \frac{Q_\theta}{2\sqrt{Q}}d\theta$$

  Coordinate basis:
  $$d\theta^1 = \frac{Q_r}{2\sqrt{Q}},\underbrace{dr \wedge dr}{=0} + \frac{Q\theta}{2\sqrt{Q}},d\theta \wedge dr$$

  $$d\theta^1 = \frac{Q_\theta}{2\sqrt{Q}},d\theta \wedge dr$$

  Convert:
  $$d\theta \wedge dr = \frac{\theta^2}{r\sqrt{K}} \wedge \frac{\theta^1}{\sqrt{Q}} = \frac{1}{r\sqrt{QK}},\theta^2 \wedge \theta^1 =
  -\frac{1}{r\sqrt{QK}},\theta^1 \wedge \theta^2$$

  Plug in:
  $$d\theta^1 = \frac{Q_\theta}{2\sqrt{Q}}\left(-\frac{1}{r\sqrt{QK}}\right)\theta^1 \wedge \theta^2$$

  $$\boxed{d\theta^1 = -\frac{Q_\theta}{2rQ\sqrt{K}};\theta^1 \wedge \theta^2}$$

  ---
  3. Computation of $d\theta^2 = d(r\sqrt{K},d\theta)$

  Leibniz rule:
  $$d\theta^2 = d(r\sqrt{K}) \wedge d\theta + r\sqrt{K} \wedge \underbrace{d(d\theta)}_{=0}$$

  Scalar differential:
  $$\frac{\partial(r\sqrt{K})}{\partial r} = \sqrt{K} + r\cdot\frac{K_r}{2\sqrt{K}} = \frac{2K + rK_r}{2\sqrt{K}}$$

  $$\frac{\partial(r\sqrt{K})}{\partial\theta} = r\cdot\frac{K_\theta}{2\sqrt{K}} = \frac{rK_\theta}{2\sqrt{K}}$$

  $$d(r\sqrt{K}) = \frac{2K + rK_r}{2\sqrt{K}},dr + \frac{rK_\theta}{2\sqrt{K}},d\theta$$

  Coordinate basis:
  $$d\theta^2 = \frac{2K + rK_r}{2\sqrt{K}},dr \wedge d\theta + \frac{rK_\theta}{2\sqrt{K}},\underbrace{d\theta \wedge d\theta}_{=0}$$

  $$d\theta^2 = \frac{2K + rK_r}{2\sqrt{K}},dr \wedge d\theta$$

  Convert:
  $$dr \wedge d\theta = \frac{\theta^1}{\sqrt{Q}} \wedge \frac{\theta^2}{r\sqrt{K}} = \frac{1}{r\sqrt{QK}},\theta^1 \wedge \theta^2$$

  Plug in:
  $$d\theta^2 = \frac{2K + rK_r}{2\sqrt{K}} \cdot \frac{1}{r\sqrt{QK}},\theta^1 \wedge \theta^2$$

  $$\boxed{d\theta^2 = \frac{2K + rK_r}{2rK\sqrt{Q}};\theta^1 \wedge \theta^2}$$

  ---
  4. Computation of $d\theta^3 = d!\left(r\sqrt{K}\sin\theta,(d\phi - \Omega dt)\right)$

  Define:
  $$f = r\sqrt{K}\sin\theta,\qquad \alpha = d\phi - \Omega dt$$

  Leibniz rule:
  $$d\theta^3 = df \wedge \alpha + f,d\alpha$$

  4a. Scalar differential $df$

  $$\frac{\partial f}{\partial r} = \sin\theta \cdot \frac{2K + rK_r}{2\sqrt{K}}$$

  $$\frac{\partial f}{\partial\theta} = r\left[\frac{K_\theta}{2\sqrt{K}}\sin\theta + \sqrt{K}\cos\theta\right] =
  \frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta$$

  $$df = \sin\theta,\frac{2K + rK_r}{2\sqrt{K}},dr + \left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)d\theta$$

  4b. Exterior derivative $d\alpha$

  $$d\alpha = d(d\phi) - d(\Omega dt) = 0 - d(\Omega dt)$$

  $$d(\Omega dt) = d\Omega \wedge dt + \Omega \wedge \underbrace{d(dt)}_{=0} = d\Omega \wedge dt$$

  $$d\Omega = \Omega_r,dr + \Omega_\theta,d\theta$$

  $$d\alpha = -\Omega_r,dr \wedge dt - \Omega_\theta,d\theta \wedge dt$$

  4c. First term: $df \wedge \alpha$

  $$\alpha = d\phi - \Omega dt$$

  $$df \wedge \alpha = \sin\theta,\frac{2K + rK_r}{2\sqrt{K}},dr \wedge d\phi
                  - \sin\theta,\frac{2K + rK_r}{2\sqrt{K}},\Omega,dr \wedge dt$$
  $$              + \left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)d\theta \wedge d\phi$$
  $$              - \left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\Omega,d\theta \wedge dt$$

  4d. Second term: $f,d\alpha$

  $$f,d\alpha = r\sqrt{K}\sin\theta\left(-\Omega_r,dr \wedge dt - \Omega_\theta,d\theta \wedge dt\right)$$
  $$= -r\sqrt{K}\sin\theta,\Omega_r,dr \wedge dt - r\sqrt{K}\sin\theta,\Omega_\theta,d\theta \wedge dt$$

  4e. Sum in coordinate basis

  $$d\theta^3 = \underbrace{\sin\theta,\frac{2K + rK_r}{2\sqrt{K}}}{A_1},dr \wedge d\phi$$
  $$           -\underbrace{\left[\sin\theta,\frac{2K + rK_r}{2\sqrt{K}}\Omega + r\sqrt{K}\sin\theta,\Omega_r\right]}{B_1},dr \wedge dt$$
  $$           +\underbrace{\left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)}{C_1},d\theta \wedge d\phi$$
  $$           -\underbrace{\left[\left(\frac{r\sin\theta,K\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\Omega +
  r\sqrt{K}\sin\theta,\Omega_\theta\right]}_{D_1},d\theta \wedge dt$$

  4f. Convert coordinate wedge products to tetrad basis

  First, expressions for $d\phi$:
  $$d\phi = \Omega,dt + \frac{\theta^3}{r\sqrt{K}\sin\theta} = \frac{\Omega}{\sqrt{H}}\theta^0 + \frac{\theta^3}{r\sqrt{K}\sin\theta}$$

  Now the four coordinate wedge products:

  $dr \wedge dt$:
  $$dr \wedge dt = \frac{\theta^1}{\sqrt{Q}} \wedge \frac{\theta^0}{\sqrt{H}} = -\frac{1}{\sqrt{HQ}},\theta^0 \wedge \theta^1$$

  $dr \wedge d\phi$:
  $$dr \wedge d\phi = \frac{\theta^1}{\sqrt{Q}} \wedge \left(\frac{\Omega}{\sqrt{H}}\theta^0 + \frac{\theta^3}{r\sqrt{K}\sin\theta}\right)$$
  $$= \frac{\Omega}{\sqrt{HQ}},\theta^1 \wedge \theta^0 + \frac{1}{r\sqrt{KQ}\sin\theta},\theta^1 \wedge \theta^3$$
  $$= -\frac{\Omega}{\sqrt{HQ}},\theta^0 \wedge \theta^1 + \frac{1}{r\sqrt{KQ}\sin\theta},\theta^1 \wedge \theta^3$$

  $d\theta \wedge dt$:
  $$d\theta \wedge dt = \frac{\theta^2}{r\sqrt{K}} \wedge \frac{\theta^0}{\sqrt{H}} = -\frac{1}{r\sqrt{HK}},\theta^0 \wedge \theta^2$$

  $d\theta \wedge d\phi$:
  $$d\theta \wedge d\phi = \frac{\theta^2}{r\sqrt{K}} \wedge \left(\frac{\Omega}{\sqrt{H}}\theta^0 + \frac{\theta^3}{r\sqrt{K}\sin\theta}\right)$$
  $$= \frac{\Omega}{r\sqrt{HK}},\theta^2 \wedge \theta^0 + \frac{1}{r^2K\sin\theta},\theta^2 \wedge \theta^3$$
  $$= -\frac{\Omega}{r\sqrt{HK}},\theta^0 \wedge \theta^2 + \frac{1}{r^2K\sin\theta},\theta^2 \wedge \theta^3$$

  4g. Substitute each term

  Term A ($dr \wedge d\phi$):
  $$A = \sin\theta,\frac{2K + rK_r}{2\sqrt{K}}\left(-\frac{\Omega}{\sqrt{HQ}},\theta^0 \wedge \theta^1 + \frac{1}{r\sqrt{KQ}\sin\theta},\theta^1
  \wedge \theta^3\right)$$

  $$= -\frac{\sin\theta,(2K + rK_r)\Omega}{2\sqrt{HKQ}};\theta^0 \wedge \theta^1 ;+; \frac{2K + rK_r}{2rK\sqrt{Q}};\theta^1 \wedge \theta^3$$

  Term B ($dr \wedge dt$):
  $$B = -\left[\sin\theta,\frac{2K + rK_r}{2\sqrt{K}}\Omega + r\sqrt{K}\sin\theta,\Omega_r\right]\left(-\frac{1}{\sqrt{HQ}},\theta^0 \wedge
  \theta^1\right)$$

  $$= \frac{\sin\theta,(2K + rK_r)\Omega}{2\sqrt{HKQ}};\theta^0 \wedge \theta^1 ;+; \frac{r\sqrt{K}\sin\theta,\Omega_r}{\sqrt{HQ}};\theta^0 \wedge
  \theta^1$$

  Observe: The first sub-term of A and the first sub-term of B cancel:
  $$-\frac{\sin\theta,(2K + rK_r)\Omega}{2\sqrt{HKQ}} + \frac{\sin\theta,(2K + rK_r)\Omega}{2\sqrt{HKQ}} = 0$$

  Term C ($d\theta \wedge d\phi$):
  Let $C_1 = \frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta$

  $$C = C_1\left(-\frac{\Omega}{r\sqrt{HK}},\theta^0 \wedge \theta^2 + \frac{1}{r^2K\sin\theta},\theta^2 \wedge \theta^3\right)$$

  First sub-term:
  $$-\frac{\Omega}{r\sqrt{HK}}\left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\theta^0 \wedge \theta^2$$
  $$= -\Omega\left(\frac{\sin\theta,K_\theta}{2\sqrt{HK}} + \frac{\cos\theta}{\sqrt{H}}\right)\theta^0 \wedge \theta^2$$

  Second sub-term:
  $$\frac{1}{r^2K\sin\theta}\left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\theta^2 \wedge \theta^3$$
  $$= \left(\frac{K_\theta}{2rK^{3/2}} + \frac{\cot\theta}{r\sqrt{K}}\right)\theta^2 \wedge \theta^3$$

  Term D ($d\theta \wedge dt$):
  Let $D_1 = \left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\Omega + r\sqrt{K}\sin\theta,\Omega_\theta$

  $$D = -D_1\left(-\frac{1}{r\sqrt{HK}},\theta^0 \wedge \theta^2\right) = \frac{D_1}{r\sqrt{HK}};\theta^0 \wedge \theta^2$$

  Expand:
  $$\frac{1}{r\sqrt{HK}}\left[\left(\frac{r\sin\theta,K_\theta}{2\sqrt{K}} + r\sqrt{K}\cos\theta\right)\Omega +
  r\sqrt{K}\sin\theta,\Omega_\theta\right]$$
  $$= \frac{\Omega\sin\theta,K_\theta}{2\sqrt{HK}} + \frac{\Omega\cos\theta}{\sqrt{H}} + \frac{\sin\theta,\Omega_\theta}{\sqrt{H}}$$

  Observe: The first two sub-terms of D cancel with the first sub-term of C:
  $$-\Omega\frac{\sin\theta K_\theta}{2\sqrt{HK}} + \frac{\Omega\sin\theta K_\theta}{2\sqrt{HK}} = 0$$
  $$-\frac{\Omega\cos\theta}{\sqrt{H}} + \frac{\Omega\cos\theta}{\sqrt{H}} = 0$$

  4h. Collect surviving terms

\[
    \fbox{$
        \begin{aligned} 
        d\theta^3 = &\frac{r\sqrt{K}\sin\theta,\Omega_r}{\sqrt{HQ}};\theta^0 \wedge \theta^1 \\[4pt] 
        + &\frac{\sin\theta,\Omega_\theta}{\sqrt{H}};\theta^0 \wedge \theta^2 \\[4pt]
        + &\frac{2K + rK_r}{2rK\sqrt{Q}};\theta^1 \wedge \theta^3 \\[4pt] 
        + &\left(\frac{K_\theta}{2rK^{3/2}} + \frac{\cot\theta}{r\sqrt{K}}\right)\theta^2 \wedge \theta^3
        \end{aligned}
    $}
\]

  ---
  Final Results

\[
    \fbox{$
        \begin{aligned}
        d\theta^0 &= -\frac{H_r}{2H\sqrt{Q}};\theta^0 \wedge \theta^1 ;-; \frac{H_\theta}{2rH\sqrt{K}};\theta^0 \wedge \theta^2\\[8pt]
        d\theta^1 &= -\frac{Q_\theta}{2rQ\sqrt{K}};\theta^1 \wedge \theta^2 \\[8pt]
        d\theta^2 &= \frac{2K + rK_r}{2rK\sqrt{Q}};\theta^1 \wedge \theta^2 \\[8pt]
        d\theta^3 &= \frac{r\sqrt{K}\sin\theta,\Omega_r}{\sqrt{HQ}};\theta^0 \wedge \theta^1 ;+;
        \frac{\sin\theta,\Omega_\theta}{\sqrt{H}};\theta^0 \wedge \theta^2 \ &\quad + \frac{2K + rK_r}{2rK\sqrt{Q}};\theta^1 \wedge \theta^3 ;+;
        \frac{1}{r\sqrt{K}}\left(\frac{K_\theta}{2K} + \cot\theta\right)\theta^2 \wedge \theta^3 
        \end{aligned}
    $}
\]

  Notation: $H_r = \partial_r H$, $H_\theta = \partial_\theta H$, and similarly for $Q$, $K$, $\Omega$.

  Key observations:
  - $d\theta^1$ and $d\theta^2$ involve only $\theta^1 \wedge \theta^2$, with no $r$-derivative of $Q$ in $d\theta^1$ and no $\theta$-derivative of
  $K$ in $d\theta^2$
  - The $\theta^0 \wedge \theta^1$ terms in $d\theta^3$ originate purely from $\Omega_r$ (the $\Omega$ terms from the $dr \wedge d\phi$ and $dr
  \wedge dt$ contributions cancel exactly)
  - The $\theta^0 \wedge \theta^2$ terms in $d\theta^3$ originate purely from $\Omega_\theta$ (similar cancellation)
  - The $\theta^1 \wedge \theta^3$ coefficient in $d\theta^3$ equals the $\theta^1 \wedge \theta^2$ coefficient in $d\theta^2$

Levi-Civita Connection One-Forms via Cartan's First Structure Equation

  Previously established exterior derivatives

  For brevity, define the following coefficient functions:

  $$\begin{aligned}
  \alpha &= \frac{H_r}{2H\sqrt{Q}}, &
  \beta &= \frac{H_\theta}{2rH\sqrt{K}}, \[4pt]
  \gamma &= \frac{Q_\theta}{2rQ\sqrt{K}}, &
  \delta &= \frac{2K + rK_r}{2rK\sqrt{Q}}, \[4pt]
  A &= \frac{r\sqrt{K}\sin\theta,\Omega_r}{\sqrt{HQ}}, &
  B &= \frac{\sin\theta,\Omega_\theta}{\sqrt{H}}, \[4pt]
  D &= \frac{K_\theta}{2rK^{3/2}} + \frac{\cot\theta}{r\sqrt{K}}.
  \end{aligned}$$

  The exterior derivatives are:

  $$\begin{aligned}
  d\theta^0 &= -\alpha,\theta^0\wedge\theta^1 - \beta,\theta^0\wedge\theta^2, \
  d\theta^1 &= -\gamma,\theta^1\wedge\theta^2, \
  d\theta^2 &= \delta,\theta^1\wedge\theta^2, \
  d\theta^3 &= A,\theta^0\wedge\theta^1 + B,\theta^0\wedge\theta^2 + \delta,\theta^1\wedge\theta^3 + D,\theta^2\wedge\theta^3.
  \end{aligned}$$

  These are taken as given — no rederivation.

  ---
  Cartan's first structure equations

  For the torsion-free Levi-Civita connection:

  $$\boxed{d\theta^a + \omega^a{}_b \wedge \theta^b = 0}$$

  with metric compatibility (antisymmetry):

  $$\boxed{\omega_{ab} = -\omega_{ba}}$$

  In the orthonormal frame with $\eta_{ab} = \operatorname{diag}(-1,+1,+1,+1)$, this implies:

  $$\omega^0{}_1 = \omega^1{}_0,\quad \omega^0{}_2 = \omega^2{}_0,\quad \omega^0{}_3 = \omega^3{}_0,$$
  $$\omega^2{}_1 = -\omega^1{}_2,\quad \omega^3{}_1 = -\omega^1{}_3,\quad \omega^3{}_2 = -\omega^2{}_3,$$
  $$\omega^0{}_0 = \omega^1{}_1 = \omega^2{}_2 = \omega^3{}_3 = 0.$$

  The six independent connection one-forms are:

  $$\omega^0{}_1,; \omega^0{}_2,; \omega^0{}_3,; \omega^1{}_2,; \omega^1{}_3,; \omega^2{}_3.$$

  ---
  Ansatz

  Each connection one-form is expanded in the orthonormal basis:

  $$\omega^a{}_b = C^a{}_b{}^0,\theta^0 + C^a{}_b{}^1,\theta^1 + C^a{}_b{}^2,\theta^2 + C^a{}_b{}^3,\theta^3.$$

  The Cartan equations are:

  $$\begin{aligned}
  \text{(E0)}\quad & d\theta^0 + \omega^0{}_1\wedge\theta^1 + \omega^0{}_2\wedge\theta^2 + \omega^0{}_3\wedge\theta^3 = 0, \
  \text{(E1)}\quad & d\theta^1 + \omega^0{}_1\wedge\theta^0 + \omega^1{}_2\wedge\theta^2 + \omega^1{}_3\wedge\theta^3 = 0, \
  \text{(E2)}\quad & d\theta^2 + \omega^0{}_2\wedge\theta^0 - \omega^1{}_2\wedge\theta^1 + \omega^2{}_3\wedge\theta^3 = 0, \
  \text{(E3)}\quad & d\theta^3 + \omega^0{}_3\wedge\theta^0 - \omega^1{}_3\wedge\theta^1 - \omega^2{}_3\wedge\theta^2 = 0.
  \end{aligned}$$

  ---
  Solving $\omega^0{}_1$

  Ansatz: $\omega^0{}_1 = a_0\theta^0 + a_1\theta^1 + a_2\theta^2 + a_3\theta^3$.

  Cartan equation (E0):

  $$d\theta^0 + \omega^0{}_1\wedge\theta^1 + \omega^0{}_2\wedge\theta^2 + \omega^0{}_3\wedge\theta^3 = 0.$$

  Expand $\omega^0{}_1\wedge\theta^1$:

  $$\begin{aligned}
  \omega^0{}1\wedge\theta^1 &= (a_0\theta^0 + a_1\theta^1 + a_2\theta^2 + a_3\theta^3)\wedge\theta^1 \
  &= a_0,\theta^0\wedge\theta^1 + a_1,\underbrace{\theta^1\wedge\theta^1}{=0} + a_2,\theta^2\wedge\theta^1 + a_3,\theta^3\wedge\theta^1 \
  &= a_0,\theta^0\wedge\theta^1 - a_2,\theta^1\wedge\theta^2 - a_3,\theta^1\wedge\theta^3.
  \end{aligned}$$

  Expand $\omega^0{}_2\wedge\theta^2$ (with $\omega^0{}_2 = b_0\theta^0 + b_1\theta^1 + b_2\theta^2 + b_3\theta^3$):

  $$\begin{aligned}
  \omega^0{}2\wedge\theta^2 &= b_0,\theta^0\wedge\theta^2 + b_1,\theta^1\wedge\theta^2 + b_2,\underbrace{\theta^2\wedge\theta^2}{=0} +
  b_3,\theta^3\wedge\theta^2 \
  &= b_0,\theta^0\wedge\theta^2 + b_1,\theta^1\wedge\theta^2 - b_3,\theta^2\wedge\theta^3.
  \end{aligned}$$

  Expand $\omega^0{}_3\wedge\theta^3$ (with $\omega^0{}_3 = c_0\theta^0 + c_1\theta^1 + c_2\theta^2 + c_3\theta^3$):

  $$\begin{aligned}
  \omega^0{}3\wedge\theta^3 &= c_0,\theta^0\wedge\theta^3 + c_1,\theta^1\wedge\theta^3 + c_2,\theta^2\wedge\theta^3 +
  c_3,\underbrace{\theta^3\wedge\theta^3}{=0} \
  &= c_0,\theta^0\wedge\theta^3 + c_1,\theta^1\wedge\theta^3 + c_2,\theta^2\wedge\theta^3.
  \end{aligned}$$

  Collect all terms from (E0):

  $$\begin{aligned}
  &d\theta^0 + \omega^0{}_1\wedge\theta^1 + \omega^0{}_2\wedge\theta^2 + \omega^0{}_3\wedge\theta^3 = \
  &\quad (-\alpha + a_0),\theta^0\wedge\theta^1 \
  &\quad + (-\beta + b_0),\theta^0\wedge\theta^2 \
  &\quad + c_0,\theta^0\wedge\theta^3 \
  &\quad + (-a_2 + b_1),\theta^1\wedge\theta^2 \
  &\quad + (-a_3 + c_1),\theta^1\wedge\theta^3 \
  &\quad + (-b_3 + c_2),\theta^2\wedge\theta^3 = 0.
  \end{aligned}$$

  Since the six basis two-forms are linearly independent, each coefficient must vanish:

  $$\boxed{a_0 = \alpha},\quad \boxed{b_0 = \beta},\quad \boxed{c_0 = 0},$$
  $$\boxed{b_1 = a_2},\quad \boxed{c_1 = a_3},\quad \boxed{c_2 = b_3}.$$

  Now we turn to (E1) to constrain $a_1, a_2, a_3$ further.

  Cartan equation (E1):

  $$d\theta^1 + \omega^0{}_1\wedge\theta^0 + \omega^1{}_2\wedge\theta^2 + \omega^1{}_3\wedge\theta^3 = 0.$$

  Expand $\omega^0{}_1\wedge\theta^0$:

  $$\omega^0{}1\wedge\theta^0 = a_0,\underbrace{\theta^0\wedge\theta^0}{=0} + a_1,\theta^1\wedge\theta^0 + a_2,\theta^2\wedge\theta^0 +
  a_3,\theta^3\wedge\theta^0
  = -a_1,\theta^0\wedge\theta^1 - a_2,\theta^0\wedge\theta^2 - a_3,\theta^0\wedge\theta^3.$$

  Expand $\omega^1{}_2\wedge\theta^2$ (with $\omega^1{}_2 = p_0\theta^0 + p_1\theta^1 + p_2\theta^2 + p_3\theta^3$):

  $$\omega^1{}_2\wedge\theta^2 = p_0,\theta^0\wedge\theta^2 + p_1,\theta^1\wedge\theta^2 + p_3,\theta^3\wedge\theta^2
  = p_0,\theta^0\wedge\theta^2 + p_1,\theta^1\wedge\theta^2 - p_3,\theta^2\wedge\theta^3.$$

  Expand $\omega^1{}_3\wedge\theta^3$ (with $\omega^1{}_3 = q_0\theta^0 + q_1\theta^1 + q_2\theta^2 + q_3\theta^3$):

  $$\omega^1{}_3\wedge\theta^3 = q_0,\theta^0\wedge\theta^3 + q_1,\theta^1\wedge\theta^3 + q_2,\theta^2\wedge\theta^3.$$

  Collect all terms from (E1):

  $$\begin{aligned}
  &(-\gamma,\theta^1\wedge\theta^2) \
  &\quad + (-a_1),\theta^0\wedge\theta^1 \
  &\quad + (-a_2 + p_0),\theta^0\wedge\theta^2 \
  &\quad + (-a_3 + q_0),\theta^0\wedge\theta^3 \
  &\quad + (-\gamma + p_1),\theta^1\wedge\theta^2 \
  &\quad + q_1,\theta^1\wedge\theta^3 \
  &\quad + (-p_3 + q_2),\theta^2\wedge\theta^3 = 0.
  \end{aligned}$$

  Each coefficient must vanish:

  $$\boxed{a_1 = 0},\quad \boxed{p_0 = a_2},\quad \boxed{q_0 = a_3},$$
  $$\boxed{p_1 = \gamma},\quad \boxed{q_1 = 0},\quad \boxed{q_2 = p_3}.$$

  Cartan equation (E2):

  $$d\theta^2 + \omega^0{}_2\wedge\theta^0 - \omega^1{}_2\wedge\theta^1 + \omega^2{}_3\wedge\theta^3 = 0.$$

Curvature Two-Forms via Cartan's Second Structure Equation

  Preliminaries

  For brevity, we define the following coefficient functions (taken from the previous steps):

  $$\boxed{\begin{aligned}
  \alpha &= \frac{H_r}{2H\sqrt{Q}}, &
  \beta &= \frac{H_\theta}{2rH\sqrt{K}}, \[6pt]
  \gamma &= \frac{Q_\theta}{2rQ\sqrt{K}}, &
  \delta &= \frac{2K + rK_r}{2rK\sqrt{Q}}, \[6pt]
  A &= \frac{r\sqrt{K}\sin\theta,\Omega_r}{\sqrt{HQ}}, &
  B &= \frac{\sin\theta,\Omega_\theta}{\sqrt{H}}, \[6pt]
  D &= \frac{1}{r\sqrt{K}}!\left(\frac{K_\theta}{2K} + \cot\theta\right).
  \end{aligned}}$$

  The orthonormal coframe exterior derivatives are:

  $$\begin{aligned}
  d\theta^0 &= -\alpha,\theta^0!\wedge!\theta^1 - \beta,\theta^0!\wedge!\theta^2, \
  d\theta^1 &= -\gamma,\theta^1!\wedge!\theta^2, \
  d\theta^2 &= \delta,\theta^1!\wedge!\theta^2, \
  d\theta^3 &= A,\theta^0!\wedge!\theta^1 + B,\theta^0!\wedge!\theta^2 + \delta,\theta^1!\wedge!\theta^3 + D,\theta^2!\wedge!\theta^3.
  \end{aligned}$$

  The six independent connection one-forms are:

  $$\boxed{\begin{aligned}
  \omega^0{}_1 &= \alpha,\theta^0 + \tfrac{A}{2},\theta^3, \[4pt]
  \omega^0{}_2 &= \beta,\theta^0 + \tfrac{B}{2},\theta^3, \[4pt]
  \omega^0{}_3 &= \tfrac{A}{2},\theta^1 + \tfrac{B}{2},\theta^2, \[4pt]
  \omega^1{}_2 &= \gamma,\theta^1 - \delta,\theta^2, \[4pt]
  \omega^1{}_3 &= \tfrac{A}{2},\theta^0 - \delta,\theta^3, \[4pt]
  \omega^2{}_3 &= \tfrac{B}{2},\theta^0 - D,\theta^3.
  \end{aligned}}$$

  For any scalar function $f(r,\theta)$, we write its exterior derivative as:

  $$df = \frac{f_r}{\sqrt{Q}},\theta^1 + \frac{f_\theta}{r\sqrt{K}},\theta^2 \equiv f_1,\theta^1 + f_2,\theta^2.$$

  The $\theta^0$ and $\theta^3$ components of $df$ are always zero since $f$ depends only on $(r,\theta)$.

  ---
  For reference, the non-zero components we will need:

  $$\boxed{\begin{aligned}
  d\alpha &= \alpha_1\theta^1 + \alpha_2\theta^2, &
  d\beta &= \beta_1\theta^1 + \beta_2\theta^2, \
  d\gamma &= \gamma_1\theta^1 + \gamma_2\theta^2, &
  d\delta &= \delta_1\theta^1 + \delta_2\theta^2, \[4pt]
  d!\left(\tfrac{A}{2}\right) &= \tfrac{A_1}{2}\theta^1 + \tfrac{A_2}{2}\theta^2, &
  d!\left(\tfrac{B}{2}\right) &= \tfrac{B_1}{2}\theta^1 + \tfrac{B_2}{2}\theta^2, \[4pt]
  dD &= D_1\theta^1 + D_2\theta^2.
  \end{aligned}}$$

  where

  $$\boxed{\begin{aligned}
  \alpha_1 = \frac{\alpha_r}{\sqrt{Q}},\quad \alpha_2 = \frac{\alpha_\theta}{r\sqrt{K}},\quad
  A_1 = \frac{A_r}{\sqrt{Q}},\quad A_2 = \frac{A_\theta}{r\sqrt{K}},\quad \text{etc.}
  \end{aligned}}$$

  ---
  Cartan's Second Structure Equation

  $$\boxed{\Omega^a{}_b = d\omega^a{}_b + \omega^a{}_c \wedge \omega^c{}_b}$$

  Metric compatibility: $\Omega_{ab} = -\Omega_{ba}$.

  The six independent curvature two-forms are:
  $\Omega^0{}_1,; \Omega^0{}_2,; \Omega^0{}_3,; \Omega^1{}_2,; \Omega^1{}_3,; \Omega^2{}_3$.

  ---
  1. $\Omega^0{}_1$

  1a. Connection form

  $$\omega^0{}_1 = \alpha,\theta^0 + \frac{A}{2},\theta^3.$$

  1b. Exterior derivative

  $$\begin{aligned}
  d\omega^0{}_1 &= d\alpha \wedge \theta^0 + \alpha,d\theta^0 + d!\left(\frac{A}{2}\right) \wedge \theta^3 + \frac{A}{2},d\theta^3.
  \end{aligned}$$

  Term 1: $d\alpha \wedge \theta^0$:

  $$\begin{aligned}
  d\alpha \wedge \theta^0 &= (\alpha_1\theta^1 + \alpha_2\theta^2) \wedge \theta^0 \
  &= \alpha_1,\theta^1!\wedge!\theta^0 + \alpha_2,\theta^2!\wedge!\theta^0 \
  &= -\alpha_1,\theta^0!\wedge!\theta^1 - \alpha_2,\theta^0!\wedge!\theta^2.
  \end{aligned}$$

  Term 2: $\alpha,d\theta^0$:

  $$\alpha(-\alpha,\theta^0!\wedge!\theta^1 - \beta,\theta^0!\wedge!\theta^2) = -\alpha^2,\theta^0!\wedge!\theta^1 -
  \alpha\beta,\theta^0!\wedge!\theta^2.$$

  Term 3: $d(A/2) \wedge \theta^3$:

  $$\begin{aligned}
  d!\left(\frac{A}{2}\right) \wedge \theta^3 &= \left(\frac{A_1}{2}\theta^1 + \frac{A_2}{2}\theta^2\right) \wedge \theta^3 \
  &= \frac{A_1}{2},\theta^1!\wedge!\theta^3 + \frac{A_2}{2},\theta^2!\wedge!\theta^3.
  \end{aligned}$$

  Term 4: $\frac{A}{2},d\theta^3$:

  $$\frac{A}{2}(A,\theta^0!\wedge!\theta^1 + B,\theta^0!\wedge!\theta^2 + \delta,\theta^1!\wedge!\theta^3 + D,\theta^2!\wedge!\theta^3).$$

  $$\begin{aligned}
  &= \frac{A^2}{2},\theta^0!\wedge!\theta^1 + \frac{AB}{2},\theta^0!\wedge!\theta^2 + \frac{A\delta}{2},\theta^1!\wedge!\theta^3 +
  \frac{AD}{2},\theta^2!\wedge!\theta^3.
  \end{aligned}$$

  Summing all four terms for $d\omega^0{}_1$:

  $$\boxed{\begin{aligned}
  d\omega^0{}_1 = ;&(-\alpha_1 - \alpha^2 + \tfrac{A^2}{2});\theta^0!\wedge!\theta^1 \
  +;&(-\alpha_2 - \alpha\beta + \tfrac{AB}{2});\theta^0!\wedge!\theta^2 \
  +;&\left(\tfrac{A_1}{2} + \tfrac{A\delta}{2}\right)\theta^1!\wedge!\theta^3 \
  +;&\left(\tfrac{A_2}{2} + \tfrac{AD}{2}\right)\theta^2!\wedge!\theta^3.
  \end{aligned}}$$

  1c. Wedge product $\omega^0{}_c \wedge \omega^c{}_1$

  The non-zero contributions (using $\omega_{ab} = -\omega_{ba}$):

  $$\begin{aligned}
  \omega^0{}_c \wedge \omega^c{}_1 &= \omega^0{}_0 \wedge \omega^0{}_1 + \omega^0{}_1 \wedge \omega^1{}_1 + \omega^0{}_2 \wedge \omega^2{}_1 +
  \omega^0{}_3 \wedge \omega^3{}_1 \
  &= 0 + \omega^0{}_1 \wedge 0 + \omega^0{}_2 \wedge (-\omega^1{}_2) + \omega^0{}_3 \wedge (-\omega^1{}_3) \
  &= -\omega^0{}_2 \wedge \omega^1{}_2 - \omega^0{}_3 \wedge \omega^1{}_3.
  \end{aligned}$$

  First product: $\omega^0{}_2 \wedge \omega^1{}_2$:

  $$\begin{aligned}
  \omega^0{}_2 &= \beta\theta^0 + \tfrac{B}{2}\theta^3, \
  \omega^1{}_2 &= \gamma\theta^1 - \delta\theta^2.
  \end{aligned}$$

  $$\begin{aligned}
  \omega^0{}_2 \wedge \omega^1{}_2 &= (\beta\theta^0 + \tfrac{B}{2}\theta^3) \wedge (\gamma\theta^1 - \delta\theta^2) \
  &= \beta\gamma,\theta^0!\wedge!\theta^1 - \beta\delta,\theta^0!\wedge!\theta^2 \
  &\quad + \tfrac{B\gamma}{2},\theta^3!\wedge!\theta^1 - \tfrac{B\delta}{2},\theta^3!\wedge!\theta^2.
  \end{aligned}$$

  Using $\theta^3!\wedge!\theta^1 = -\theta^1!\wedge!\theta^3$ and $\theta^3!\wedge!\theta^2 = \theta^2!\wedge!\theta^3$:

  $$\begin{aligned}
  \omega^0{}_2 \wedge \omega^1{}_2 &= \beta\gamma,\theta^0!\wedge!\theta^1 - \beta\delta,\theta^0!\wedge!\theta^2
  - \tfrac{B\gamma}{2},\theta^1!\wedge!\theta^3 - \tfrac{B\delta}{2},\theta^2!\wedge!\theta^3.
  \end{aligned}$$
